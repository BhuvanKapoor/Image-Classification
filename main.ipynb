```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
X,y = [], []
images = [img for img in os.listdir('train')]
```

```python
for img in images:
    label = img.split('.')[0]
    img_no = img.split('.')[1]
    image_path = os.path.join('train',img)
    train_img = cv2.imread(image_path)
    scaled_img = cv2.resize(train_img, (64,64))
    X.append(scaled_img.flatten())
    if label == 'cat':
        y.append(0)
    else:
        y.append(1)
    # print(img)
```

```python
img = cv2.imread('train/cat.2.jpg')
scale_img = cv2.resize(img, (64,64))
plt.imshow(scale_img)
plt.show()
```

```python
from sklearn.svm import SVC
```

```python
model = SVC(random_state=42)
model.fit(X,y)
```

```python
predictions = {
    "ID" : []
}
```

```python
X_pred = []
test_images = [img for img in os.listdir('test1')]
```

```python
for img in test_images:
    id = img.split('.')[0]
    predictions["ID"].append(id)
    image_path = os.path.join('test1',img)
    test_img = cv2.imread(image_path)
    scaled_test_img = cv2.resize(test_img, (64,64))
    X_pred.append(scaled_test_img.flatten())
```

```python
y_pred = model.predict(X_pred)
```
